## Get Playlist Items

```python
# establish playlist url from Spotify: https://open.spotify.com/playlist/37i9dQZF1DX0kbJZpiYdZl

playlist_id = '37i9dQZF1DX0kbJZpiYdZl'
```

```python
# create empty dictionary for data that you want to extract
playlist_data = {
    'song_name' : [],
    'album_name' : [],
    'artist_name' : [],
    'artist_id' : [],
    'popularity' : [],
    'song_id' : []
}

# loop through json to extract data

for item in playlist_items['items'][0:50]:
      song_name = item['track']['name']
      playlist_data['song_name'].append(song_name)
      print(song_name)

      album_name = item['track']['album']['name']
      playlist_data['album_name'].append(album_name)
      print(album_name)

      artist_name = item['track']['artists'][0]['name']
      playlist_data['artist_name'].append(artist_name)
      print(artist_name)

      artist_id = item['track']['artists'][0]['id']
      playlist_data['artist_id'].append(artist_id)
      print(artist_id)

      popularity = item['track']['popularity']
      playlist_data['popularity'].append(popularity)
      print(popularity)

      song_id = item['track']['id']
      playlist_data['song_id'].append(song_id)
      print(song_id)
    
      print('-'*75) 
```
